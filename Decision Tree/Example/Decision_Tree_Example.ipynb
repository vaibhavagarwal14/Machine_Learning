```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn import tree
```

```python
df=pd.read_csv('salaries.csv')
```

```python
df.head()
```

```python
le_c = LabelEncoder()
le_j = LabelEncoder()
le_d = LabelEncoder()
```

```python
input = df.drop('salary_more_than_100k',axis='columns')
```

```python
target = df['salary_more_than_100k']
```

```python
input['company_new']=le_c.fit_transform(input['company'])
input['job_new']=le_c.fit_transform(input['job'])
input['degree_new']=le_c.fit_transform(input['degree'])
```

```python
input
```

```python
input=input.drop(['degree','job','company'],axis='columns')
```

```python
input
```

```python
target
```

```python
model=tree.DecisionTreeClassifier()
model.fit(input,target)
```

```python
model.score(input,target)
```
