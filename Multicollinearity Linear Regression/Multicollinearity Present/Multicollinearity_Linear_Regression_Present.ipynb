Checking how the experience and age affects the salary along with the collinearity between the independent features.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import statsmodels.api as sm
```

```python
df=pd.read_csv('experience_salary.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
x=df[['YearsExperience','Age']]
y=df['Salary']
```

```python
x=sm.add_constant(x)
```

```python
x
```

```python
#Using OLS technique to fit the data
model=sm.OLS(y,x).fit()
```

```python
model.summary()
```

```python
x.iloc[:,1:].corr()
```

Thus we can see that the correlation between years of experience and age is 98%. Also, the P-score for age is > 0.05.  
This tells are that we can drop age as a feature because 98% of it will already be covered by years of experience.

This implies there is multicollinearity in this dataset.
