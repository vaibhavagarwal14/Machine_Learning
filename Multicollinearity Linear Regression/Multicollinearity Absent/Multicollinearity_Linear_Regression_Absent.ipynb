Checking how the advertisement through TV, radio and newspaper affect the sales along with the collinearity between the independent features.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import statsmodels.api as sm
```

```python
df=pd.read_csv('advertising.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
x=df[['TV','radio','newspaper']]
y=df['sales']
```

```python
x=sm.add_constant(x)
```

```python
x
```

```python
#Using OLS technique to fit the data
model=sm.OLS(y,x).fit()
```

```python
model.summary()
```

```python
x.iloc[:,1:].corr()
```

Thus we can see that all correlations are less than 0.5 and P-score is less than 0.05 (except newspaper).  
The coefficient of newspaper is -ve which means that with increase in expenditure on newspaper, the sale decrease. So we can drop it also.

This implies there is no multicollinearity in this dataset.
