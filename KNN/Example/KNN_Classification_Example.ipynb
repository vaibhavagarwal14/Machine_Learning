Performing KNN classification on Iris dataset

Loading the data

```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
```

Exploring the data

```python
iris.feature_names
```

```python
iris.target_names
```

```python
df = pd.DataFrame(iris.data,columns=iris.feature_names)
df.head()
```

```python
df['target'] = iris.target
df.head()
```

```python
df[df.target==1].head()
```

```python
df[df.target==2].head()
```

We can see that Setosa is from 0-49, Versicolor is from 50-99 and Virginica is from 100-149

```python
df['flower_name'] =df.target.apply(lambda x: iris.target_names[x])
df.head()
```

```python
df[45:55]
```

Splitting the data into three different dataframes

```python
df0 = df[:50]
df1 = df[50:100]
df2 = df[100:]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

Sepal length vs Sepal Width (Setosa vs Versicolor)

```python
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.scatter(df0['sepal length (cm)'], df0['sepal width (cm)'],color="green",marker='+')
plt.scatter(df1['sepal length (cm)'], df1['sepal width (cm)'],color="blue",marker='.')
```

Petal length vs Pepal Width (Setosa vs Versicolor)

```python
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.scatter(df0['petal length (cm)'], df0['petal width (cm)'],color="green",marker='+')
plt.scatter(df1['petal length (cm)'], df1['petal width (cm)'],color="blue",marker='.')
```

Performing Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(['target','flower_name'], axis='columns')
y = df.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
len(X_train)
```

```python
len(X_test)
```

Create KNN (K Neighrest Neighbour Classifier)

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
```

We have taken the value of k randomly as 3. However it shall be calculated using trial and error to reduce the errors.

```python
knn.fit(X_train, y_train)
```

```python
knn.score(X_test, y_test)
```

```python
knn.predict([[4.8,3.0,1.5,0.3]])
```

Checking score with k=10

```python
knn2 = KNeighborsClassifier(n_neighbors=10)
```

```python
knn2.fit(X_train, y_train)
```

```python
knn2.score(X_test, y_test)
```

We can see that as we increase the number of neighbours to be considered, the score reduces.

Plot Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
y_pred = knn.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sn
plt.figure(figsize=(7,5))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

Print classification report for precesion, recall and f1-score for each classes

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

Thus we have successfullt implemented the KNN Classification on Iris Dataset
