Implementing Gradient Descent for sample data and verifying with Linear Regression Library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import linear_model
```

```python
#We will be using y=2x+3
x=np.array([1,2,3,4,5])
y=np.array([5,7,9,11,13])
```

```python
m_curr = b_curr = 0
iterations=10
alpha=0.001
n=len(x)
```

```python
for i in range(iterations):
  y_pred = m_curr * x + b_curr
  cost=(1/n)*sum([val**2 for val in (y-y_pred)])
  md=-(2/n)*sum(x*(y-y_pred))
  bd=-(2/n)*sum(y-y_pred)
  m_curr=m_curr-alpha*md
  b_curr=b_curr-alpha*bd
  print("m - {} , b - {} , cost - {} ".format(m_curr,b_curr,cost))

#alpha=0.001, iteration=10
#We see the value of cost decreasing so we change Alpha
```

```python
alpha = 0.01
for i in range(iterations):
  y_pred = m_curr * x + b_curr
  cost=(1/n)*sum([val**2 for val in (y-y_pred)])
  md=-(2/n)*sum(x*(y-y_pred))
  bd=-(2/n)*sum(y-y_pred)
  m_curr=m_curr-alpha*md
  b_curr=b_curr-alpha*bd
  print("m - {} , b - {} , cost - {} ".format(m_curr,b_curr,cost))

#alpha=0.01, iteration=10
#We see the value of cost decreasing so we change Alpha
```

```python
alpha=0.08
for i in range(iterations):
  y_pred = m_curr * x + b_curr
  cost=(1/n)*sum([val**2 for val in (y-y_pred)])
  md=-(2/n)*sum(x*(y-y_pred))
  bd=-(2/n)*sum(y-y_pred)
  m_curr=m_curr-alpha*md
  b_curr=b_curr-alpha*bd
  print("m - {} , b - {} , cost - {} ".format(m_curr,b_curr,cost))

#alpha=0.08, iteration=10
#We see the value of cost decreasing so we fix Alpha and change number of iterations
```

```python
alpha=0.08
iterations=1000
for i in range(iterations):
  y_pred = m_curr * x + b_curr
  cost=(1/n)*sum([val**2 for val in (y-y_pred)])
  md=-(2/n)*sum(x*(y-y_pred))
  bd=-(2/n)*sum(y-y_pred)
  m_curr=m_curr-alpha*md
  b_curr=b_curr-alpha*bd
  print("m - {} , b - {} , cost - {} ".format(m_curr,b_curr,cost))

#alpha=0.08, iteration=1000
```

Thus we have reached the values of m as 2 and b as 3

```python
#Verifying using Linear Regression Model
reg=linear_model.LinearRegression()
x=x.reshape(-1,1)
y=y.reshape(-1,1)
reg.fit(x,y)
```
